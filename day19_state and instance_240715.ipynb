```python
# day19_more turtle graphics, event listener, state and multiple instances_240712
# turtle event listener

from turtle import Turtle, Screen

tim = Turtle()
screen = Screen()


def move_forwards():
    tim.fd(10)


screen.listen()
screen.onkey(key="space", fun=move_forwards)
screen.exitonclick()
```

```python
# function as inputs
# def function_a(something):
#     Do this with something
#     Then do this
#     Finally do this

# def function_b():
#     do this

def add(n1,n2):
    return n1+n2

def subtract(n1,n2):
    return n1-n2

def multiply(n1, n2):
    return n1*n2

def divide(n1,n2):
    return n1/n2

def calculator(n1, n2, func):
    return func(n1, n2)

result = calculator(2,3,add)
print(result)

# higher order function: which takes function as parameter
# In this case, calculator is higher order function.
```

```python
# Make a sketch 
from turtle import Turtle, Screen

tim = Turtle()
screen = Screen()


def move_forwards():
    tim.fd(5)

def move_backwards():
    tim.backward(5)


def turn_right():
    tim.right(5)

# teacher solution
# def turn_right():
#     new_heading = tim.heading() +10
#     tim.setheading(new_heading)


def turn_left():
    tim.left(5)


def reset():
    screen.resetscreen()
    tim.showturtle()

# teacher solution
# def clear():
#     tim.clear()
#     tim.penup()
#     tim.home()
#     tim.pendown()

screen.listen()
screen.onkey(key="w", fun=move_forwards)
screen.onkey(key="s", fun=move_backwards)
screen.onkey(key="d", fun=turn_right)
screen.onkey(key="a", fun=turn_left)
screen.onkey(key="c", fun=reset)


screen.exitonclick()
```

```python
# final_project
from turtle import Turtle, Screen
import random

is_race_on = False
screen = Screen()
screen.setup(width=500, height=400)
user_bet = screen.textinput(title="Make your bet", prompt="Which turtle will win the race? Enter a color: ")
colors = ["red", "orange", "yellow", "green", "blue", "purple"]

# tim = Turtle(shape="turtle")
# tim.color("red")
# tim.penup()
# tim.goto(x=-200, y=-100)

# Q. how can we create objects for one class in for loop?
# A. we can ues list.
# https://stackoverflow.com/questions/21598872/how-to-create-multiple-class-objects-with-a-loop-in-python

turtles = [Turtle(shape="turtle") for i in range(6)]
for t in range(6):
    turtles[t].color(colors[t])
    turtles[t].penup()
    turtles[t].goto(x=-200, y=-75+30*t)

# teacher solution
# y_position = [-70, -40, -10, 20, 50, 80]
# for turtle_index in range(0,6):
#     tim = Turtle(shape="turtle")
#     tim.color(colors[turtle_index])
#     tim.penup()
#     tim.goto(x=-230, y=y_position[turtle_index])

if user_bet:
    is_race_on = True

while is_race_on:
    for tt in range(6):
        if turtles[tt].xcor() > 230:
            is_race_on = False
            winner = turtles[tt].pencolor()
            if winner == user_bet:
                print(f"You've won! The {winner} turtle is the winner!")
            else:
                print(f"You've lost! The {winner} turtle is the winner!")

        random_distance = random.randint(0,10)
        turtles[tt].forward(random_distance)


screen.exitonclick()
```

```python
# How to modify variables with global scope
# Modifying Global scope
enemies = "Skeleton" # global scope

def increase_enemies_1():
    enimies = "Zombie" # local scope
    print(f"enemies inside function: {enimies}")

increase_enemies_1()
print(f"enemies outside function: {enemies}\n")
# it is terrible to name same name to the local and global variables

enemies = 1 # global scope

def increase_enemies_2():
    global enemies # we need to say it
    enemies += 1
    print(f"enemies inside function: {enimies}")

increase_enemies_2()
print(f"enemies outside function: {enemies}\n")
# it is confusing and prone to create bugs to use global in local
# it is not recommended to use global in local

# instead we can use return
enemies = 1 # global scope

def increase_enemies_3():
    print(f"enemies inside function: {enimies}")
    return enemies +1

enemies = increase_enemies_3()
print(f"enemies outside function: {enemies}")
```

```python
# final solution _teacher solution
from random import randint
from art import logo

EASY_LEVEL_TURNS = 10
HARD_LEVEL_TURNS = 5

# Function to check user'r guess against actual answer.
def check_answer(guess, answer, turns):
    """checks answer against guess. Returns the number of turns remaining."""
    if guess > answer:
        print("Too high.")
        return turns -1
    elif guess < answer:
        print("Too low.")
        return turns -1
    else:
        print(f"You got it! The answer was {answer}")

#Make function to set difficulty.
def set_difficulty():
    level = input("Choose a difficulty. Type 'easy' or 'hard': ")
    if level == "easy":
        return EASY_LEVEL_TURNS
    else:
        return HARD_LEVEL_TURNS

def game():
    print(logo)
    # choosing a random number between 1 and 100.
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")
    answer = randint(1,100)
    print(f"Pssst, the correst answer is {answer}")
    
    turns = set_difficulty()
    # Repeat the guessing functionality if they get it wrong.
    guess = 0
    while guess != answer:
        print(f"You have {turns} attempts reamining to guess the number.")
        # Let the user guess a number.
        guess = int(input("Make a guess: "))
        
        #Track the number of turns and reduce by 1 if they get it wrong.
        turns = check_answer(guess, answer, turns)
        if turns == 0:
            print("You've run out of guesses, you lose.")
            return
        elif guess != answer:
            print("Guess again.")

game()

# reflection: I could put left_chance varaible as input parameter of comparing_num function 
# I set the condition of while as left_chance, and teacher set the condition of while as 'guess != answer'. When we think about the purpose of the project it would be more intuitive to set the teacher's conditon as the condtion of while loop.
# Next time, let's think about the possibility to make the whole run of the game to be a function.
```

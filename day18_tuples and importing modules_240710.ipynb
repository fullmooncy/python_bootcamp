```python
# Draw a random walk _my solution
from turtle import Turtle, Screen
import random


timmy = Turtle()
timmy.shape("turtle")
timmy.speed(0)
timmy.width(5)

screen = Screen()
screen.colormode(255)

angle = [0, 90, 180, 270]


def random_direction():
    timmy.right(random.choice(angle))


def random_color():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    timmy.color(r, g, b)


for _ in range(300):
    random_direction()
    random_color()
    timmy.fd(20)

screen.exitonclick()


# Draw a random walk _teacher solution
import turtle as t
tim = t.Turtle()
colors = ["red", "green", "blue"]
directions = [0, 90, 180, 270]
tim.pensize(15)
tim.speed(0)

for _ in range(200):
    tim.color(random.choice(colors))
    tim.fd(30)
    tim.setheading(random.choice(directions))

screen.exitonclick()
```

```python
# final project_hirst spot _my solution
import colorgram
import random
import turtle
from turtle import Turtle, Screen
turtle.colormode(255)
screen = Screen()

timmy = Turtle()
timmy.shape("turtle")
timmy.color("green4")
timmy.speed(0)

# step1. color extraction from image
num_of_colors = 15
colors = colorgram.extract('hirst_img2.jpg', num_of_colors)
color_list = []
for i in range(num_of_colors):
    rgb = colors[i].rgb
    color_list.append(rgb)
# print(color_list[0])
# timmy.color(color_list[0])
# timmy.fd(100)

# step2. draw the picture

# set the start coordinate
timmy.penup()
start_x = -200
start_y = -200
timmy.setpos(start_x,start_y)
distance = 50


row_num = int(input("row num: "))
col_num = int(input("col num: "))


def row_dot(n):
    for _ in range(n-1):
        timmy.penup()
        timmy.fd(distance)
        timmy.pendown()
        timmy.dot(18, random.choice(color_list)) #size >= 1

def draw_spot(rn, cn):
    global start_x
    global start_y
    for _ in range(cn):
        timmy.dot(18, random.choice(color_list))
        row_dot(rn)
        timmy.penup()
        start_y += distance
        timmy.setpos(start_x,start_y) #UnboundLocalError: local variable 'start_y' referenced before assignment > so, put global start_x and global start_y

draw_spot(row_num,col_num)


screen.exitonclick()
```
